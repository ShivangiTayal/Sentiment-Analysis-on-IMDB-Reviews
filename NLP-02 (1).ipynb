```python
import re

# Define the preprocessing function
def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Remove special characters & numbers (optional)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
df['review'] = df['review'].apply(clean_text)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
df['review'] = df['review'].apply(remove_stopwords) 
```

```python
df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
y = []
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z =y[:]
    y.clear()
    return z
```

```python
df['review'] = df['review'].apply(stem_words) 
```

```python
df
```

```python
def join_back(text):
    return " ".join(text)
```

```python
df['review'] = df['review'].apply(join_back) 
```
